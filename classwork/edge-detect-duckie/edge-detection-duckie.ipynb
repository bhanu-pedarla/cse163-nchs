````python
import imageio
import matplotlib.pyplot as plt
import numpy as np

# Write your function here!
def edge_detect(image_file: str) -> None:
    # Define the kernel
    image_array = imageio.imread(image_file)
    kernel_sobel_x = np.array([[-1, 0, 1], 
                               [-2, 0, 2],
                               [-1, 0, 1]])
    kernel_sobel_y = np.rot90(kernel_sobel_x)
    

    r = image_array[:, :, 0].astype(float)
    g = image_array[:, :, 1].astype(float)
    b = image_array[:, :, 2].astype(float)

    image_array_gray = 0.299*r + 0.587*g + 0.114*b

    image_array_gray = np.clip(image_array_gray, 0, 255).astype(np.uint8)

    image_array_gray_padded = np.pad(image_array_gray, 1, mode = 'edge')

    image_h, image_w = image_array_gray.shape

    out_x = np.zeros((image_h, image_w), dtype = float)
    out_y = np.zeros((image_h, image_w), dtype = float)

    for ih in range(1, image_h + 1):
        for iw in range(1, image_w + 1):
            subimage = image_array_gray_padded[ih-1:ih+2, iw-1:iw+2]

            out_x[ih-1, iw-1] = np.sum(subimage * kernel_sobel_x)
            out_y[ih-1, iw-1] = np.sum(subimage * kernel_sobel_y)

    out_x = np.where(out_x > 127, 255, 0).astype(np.uint8)
    out_y = np.where(out_y > 127, 255, 0).astype(np.uint8)
    
    out_sum = out_x + out_y

    if np.max(out_sum) > 0:
        gradient_normed = (out_sum / np.max(out_sum)) * 255
    else:
        gradient_normed = out_sum

    gradient_normed = np.clip(gradient_normed, 0 , 255)

    plt.imshow(gradient_normed, cmap = 'gray')
    plt.show()

    
edge_detect("duck.jpg")
````
